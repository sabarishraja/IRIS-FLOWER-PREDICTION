# TRAIN TEST SPLITTING

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(iris_dataset['data'],iris_dataset['target'],test_size=0.2,random_state=0)
```

## BUILDING THE MODEL


**AS THIS IS A CLASIIFICATION PROBLEM, WE USE K-NEAREST NEIGHBOURS ALG. TO TRAIN THE MODEL**

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,Y_train)
```

# MAKING PREDICTIONS

import numpy as np
X_new = np.array([[5,2.9,1,0.2]])
print(X_new.shape)

```python
prediction = knn.predict(X_new)
print("Prediction:",prediction)
print(iris_dataset['target_names'][prediction])
```

### EVALUATING THE MODEL(FINDING THE ACCURACY OR PRECISION OF THE O/P OF MODEL

```python
#NOW WE USE THE TEST SET TO EVALUATE THE MODEL
Y_pred = knn.predict(X_test)
print("TEST SET PREDICTIONS: {}".format(Y_pred))
```

```python
print("TEST SET SCORE:{:.2f}".format(np.mean(Y_pred==Y_test)))
```

```python
print("TEST SET SCORE:", knn.score(X_test,Y_test))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(iris_dataset['data'],iris_dataset['target'],test_size=0.2,random_state=0)

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,Y_train)

print("TEST SET SCORE:", knn.score(X_test,Y_test))
```
